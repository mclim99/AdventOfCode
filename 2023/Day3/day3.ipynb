```python
# Part 1
def find_numbers_in_row(line):
    number_str = ""
    found_numbers = []
    symbol_locations = []

    # For each character in the current line, find the numbers in the line and their indexes
    character_index = 0
    while character_index < len(line):
        # Check if the character is a digit
        if line[character_index].isdigit():
            # If the character is a digit, store the first index and check if the next character is a digit.
            # Use string concatenation to get the number.
            first_digit_index = character_index
            next_index = character_index + 1
            number_str += line[character_index]
            while (next_index < len(line)) and (line[next_index].isdigit()):
                number_str += line[next_index]
                next_index += 1
            last_digit_index = next_index - 1 # Store the last index
            character_index = next_index # Start the iteration of the while loop after the recorded number
            
            # Store the numbers in the input line with their indexes.
            found_numbers.append((int(number_str), (first_digit_index, last_digit_index)))
            number_str = ""

        # Check if the character is a symbol and if the symbol's index has not been saved yet
        elif (line[character_index] != ".") and (not line[character_index].isdigit()) and (character_index not in symbol_locations):
            # Save the index of the symbol in the line
            symbol_locations.append(character_index)
            character_index += 1

        # Otherwise, continue the loop
        else:
            character_index += 1

    return found_numbers, symbol_locations

def find_symbols_in_adjacent_lines(symbol_locations, previous=[], next=[]):
    # For each character in the previous line, store the index of the symbol
    if previous:
        for i in range(len(previous)):
            if (previous[i] != ".") and (not previous[i].isdigit()) and (i not in symbol_locations):
                symbol_locations.append(i)

    # For each character in the next line, store the index of the symbol
    if next:
        for i in range(len(next)):
            if (next[i] != ".") and (not next[i].isdigit()) and (i not in symbol_locations):
                symbol_locations.append(i)

    return symbol_locations

def check_for_adjacent_symbols(line, existing_numbers, symbol_locations):
    global sum_of_part_numbers 
    passed_numbers = []
    
    # Compare the indexes of the numbers from the current line to check for adjacent symbols
    for i in range(len(existing_numbers)):
        number_first_index = existing_numbers[i][1][0]
        number_last_index = existing_numbers[i][1][1]

        # Account for symbols diagonal to the number
        decrement_digit_index = max(0, number_first_index - 1)
        increment_digit_index = min(number_last_index + 1, len(line))

        # Check if any index of the symbols is within range of a number, which shows that they're adjacent
        if any(symbol in symbol_locations for symbol in range(decrement_digit_index, increment_digit_index + 1)) and (existing_numbers[i] not in passed_numbers):
            sum_of_part_numbers += existing_numbers[i][0]
            passed_numbers.append(existing_numbers[i])

def main():
    with open("input.txt", 'r') as input_file:
        input_lines = [line.rstrip() for line in input_file.readlines()] # Store string inputs into a list
        
        global sum_of_part_numbers
        sum_of_part_numbers = 0
        for i in range(len(input_lines)):
            # Default starting values
            numbers_in_row = []

            # For the middle rows...
            if (i != 0) and (i != len(input_lines) - 1):
                previous_line = list(input_lines[i - 1])
                current_line = list(input_lines[i])
                next_line = list(input_lines[i + 1])

                # Store the indexes of the numbers in the input line and the indexes of the symbols
                numbers_in_row, symbol_index = find_numbers_in_row(current_line)

                # Find symbols in adjacent lines
                symbol_index = find_symbols_in_adjacent_lines(symbol_index, previous=previous_line, next=next_line)
            
            # For the first row...
            elif i == 0:
                # Store the current line and the next line as lists of characters
                current_line = list(input_lines[i])
                next_line = list(input_lines[i + 1])

                # Store the indexes of the numbers in the input line and the indexes of the symbols
                numbers_in_row, symbol_index = find_numbers_in_row(current_line)

                # Find symbols in adjacent lines
                symbol_index = find_symbols_in_adjacent_lines(symbol_index, next=next_line)

            # For the last row...    
            else:
                # Store the previous line and current line as lists of characters
                previous_line = list(input_lines[i - 1])
                current_line = list(input_lines[i])

                # Store the indexes of the numbers in the input line and the indexes of the symbols
                numbers_in_row, symbol_index = find_numbers_in_row(current_line)

                # Find symbols in adjacent lines
                symbol_index = find_symbols_in_adjacent_lines(symbol_index, previous=previous_line)

            # Compare the indexes of the numbers from the current line to check for adjacent symbols
            check_for_adjacent_symbols(current_line, numbers_in_row, symbol_index)

        # Print the sum of the part numbers
        print(sum_of_part_numbers)

if __name__ == "__main__":
    main()
```

```python
# Part 2
def find_gears_in_row(line):
    gear_index = [] # Store the index of all the gears in a row
    
    # For each character in the row...
    for i in range(len(line)):
        # If the character is a gear (asterisk), store the index in the list
        if line[i] == "*":
            gear_index.append(i)

    return gear_index

def find_numbers_in_row(line):
    number_str = ""
    found_numbers = []

    # For each character in the current line, find the numbers in the line and their indexes
    character_index = 0
    while character_index < len(line):
        # Check if the character is a digit
        if line[character_index].isdigit():
            # If the character is a digit, store the first index and check if the next character is a digit.
            # Use string concatenation to get the number.
            first_digit_index = character_index
            next_index = character_index + 1
            number_str += line[character_index]
            while (next_index < len(line)) and (line[next_index].isdigit()):
                number_str += line[next_index]
                next_index += 1
            last_digit_index = next_index - 1 # Store the last index
            character_index = next_index # Start the iteration of the while loop after the recorded number
            
            # Store the numbers in the input line with their indexes.
            found_numbers.append((int(number_str), (first_digit_index, last_digit_index)))
            number_str = ""

        # Otherwise, continue the loop
        else:
            character_index += 1

    return found_numbers

def find_numbers_adjacent_to_gears(line, gear_index, existing_numbers):
    global sum_of_gear_ratios
    gear_ratio = 0

    for gear in gear_index:
        passed_numbers = []

        # Compare the indexes of the numbers from the current line to check for adjacent gears
        for i in range(len(existing_numbers)):
            number_first_index = existing_numbers[i][1][0]
            number_last_index = existing_numbers[i][1][1]

            # Account for symbols diagonal to the number
            decrement_digit_index = max(0, number_first_index - 1)
            increment_digit_index = min(number_last_index + 1, len(line))

            # Check if the index of the gear is within range of a number, which shows that they're adjacent
            if (gear in range(decrement_digit_index, increment_digit_index + 1)) and (existing_numbers[i] not in passed_numbers):
                passed_numbers.append(existing_numbers[i])

        # If only two numbers were adjacent to a gear, calculate the gear ratio and add them to the sum
        if len(passed_numbers) == 2:
            gear_ratio = passed_numbers[0][0] * passed_numbers[1][0]
            sum_of_gear_ratios += gear_ratio

def main():
    with open("input.txt", 'r') as input_file:
        input_lines = [line.rstrip() for line in input_file.readlines()] # Store string inputs into a list
        
        global sum_of_gear_ratios
        sum_of_gear_ratios = 0
        for i in range(len(input_lines)):
            # Default starting values
            gears_in_row = []

            # For the middle rows...
            if (i != 0) and (i != len(input_lines) - 1):
                previous_line = list(input_lines[i - 1])
                current_line = list(input_lines[i])
                next_line = list(input_lines[i + 1])

                # Find all the gears in the row and save their indexes
                gears_in_row = find_gears_in_row(current_line)

                # Store the numbers and their indexes from the previous, current, and next line.
                numbers_in_previous_row = find_numbers_in_row(previous_line)
                numbers_in_current_row = find_numbers_in_row(current_line)
                numbers_in_next_row = find_numbers_in_row(next_line)
                numbers_in_rows = numbers_in_previous_row + numbers_in_current_row + numbers_in_next_row
            
            # For the first row...
            elif i == 0:
                # Store the current line and the next line as lists of characters
                current_line = list(input_lines[i])
                next_line = list(input_lines[i + 1])

                # Find all the gears in the row and save their indexes
                gears_in_row = find_gears_in_row(current_line)

                # Store the numbers and their indexes from the current line and the next line.
                numbers_in_current_row = find_numbers_in_row(current_line)
                numbers_in_next_row = find_numbers_in_row(next_line)
                numbers_in_rows = numbers_in_current_row + numbers_in_next_row

            # For the last row...    
            else:
                # Store the previous line and current line as lists of characters
                previous_line = list(input_lines[i - 1])
                current_line = list(input_lines[i])

                # Find all the gears in the row and save their indexes
                gears_in_row = find_gears_in_row(current_line)

                # Store the numbers and their indexes from the current line and the previous line.
                numbers_in_previous_row = find_numbers_in_row(next_line)
                numbers_in_current_row = find_numbers_in_row(current_line)
                numbers_in_rows = numbers_in_current_row + numbers_in_previous_row

            # Check if any numbers exist adjacent to the gears.
            # If there are only two numbers adjacent to the gears, save them.
            find_numbers_adjacent_to_gears(current_line, gears_in_row, numbers_in_rows)

        # Print the sum of the part numbers
        print(sum_of_gear_ratios)

if __name__ == "__main__":
    main()
```
