```python
# Part 1
def update_data_gathered(data_gathered, line, header):
    # If the header is in the line, update the data gathered dictionary.
    if header in line:
        data_gathered[header] = True

    return data_gathered

def check_for_headers(data_gathered, line):
    # Use the helper function to update the data gathered dictionary of booleans
    data_gathered = update_data_gathered(data_gathered, line, "seed-to-soil map")
    data_gathered = update_data_gathered(data_gathered, line, "soil-to-fertilizer map")
    data_gathered = update_data_gathered(data_gathered, line, "fertilizer-to-water map")
    data_gathered = update_data_gathered(data_gathered, line, "water-to-light map")
    data_gathered = update_data_gathered(data_gathered, line, "light-to-temperature map")
    data_gathered = update_data_gathered(data_gathered, line, "temperature-to-humidity map")
    data_gathered = update_data_gathered(data_gathered, line, "humidity-to-location map")

def label_mappings(map):
    src, dest, range_length = map[0], map[1], map[2]

    return src, dest, range_length

def main():
    with open("input.txt", 'r') as input_file:
        input_lines = [line.rstrip() for line in input_file.readlines() if line.rstrip() != ""] # Store string inputs into a list
        
        # Determine the category that each group of numbers belong to.
        # If a category is True while the next category is False, then the group of numbers still belong to the current category.
        data_gathered = {"seed-to-soil map": False, 
                         "soil-to-fertilizer map": False, 
                         "fertilizer-to-water map": False, 
                         "water-to-light map": False, 
                         "light-to-temperature map": False, 
                         "temperature-to-humidity map": False, 
                         "humidity-to-location map": False}
        
        #Use lists to store the mapping of each category
        seed_to_soil = []
        soil_to_fertililzer = []
        fertilizer_to_water = []
        water_to_light = []
        light_to_temperature = []
        temperature_to_humidity = []
        humidity_to_location = []
            
        # For each line in the list of inputs...
        for line in input_lines:
            # If "seeds" is in the line, add the seed numbers to a list
            if "seeds" in line:
                seeds = [int(i) for i in line.lstrip("seeds: ").split(" ")]

            # Check to see if the line has any of the headers for the mappings
            check_for_headers(data_gathered, line)

            # If the first character in the line is a digit, then the line is a group of numbers for one of the categories
            if line[0].isdigit():
                dest, src, range_length = [int(i) for i in line.split(" ")] # Get the destination, source, and range from the line
            
                # If the "seed-to-soil map" header has been passed, but not the "soil-to-fertilizer map" header,
                # then the group of numbers belong to the "seed-to-soil map".
                if data_gathered["seed-to-soil map"] and not data_gathered["soil-to-fertilizer map"]:
                    seed_to_soil.append((src, dest, range_length))
                    
                # If the "soil-to-fertilizer map" header has been passed, but not the "fertilizer-to-water map" header,
                # then the group of numbers belong to the "soil-to-fertilizer map".
                elif data_gathered["soil-to-fertilizer map"] and not data_gathered["fertilizer-to-water map"]:
                    soil_to_fertililzer.append((src, dest, range_length))

                # If the "fertilizer-to-water map" header has been passed, but not the "water-to-light map" header,
                # then the group of numbers belong to the "fertilizer-to-water map".
                elif data_gathered["fertilizer-to-water map"] and not data_gathered["water-to-light map"]:
                    fertilizer_to_water.append((src, dest, range_length))

                # If the "water-to-light map" header has been passed, but not the "light-to-temperature map" header,
                # then the group of numbers belong to the "water-to-light map".
                elif data_gathered["water-to-light map"] and not data_gathered["light-to-temperature map"]:
                    water_to_light.append((src, dest, range_length))

                # If the "light-to-temperature map" header has been passed, but not the "temperature-to-humidity map" header,
                # then the group of numbers belong to the "light-to-temperature map".
                elif data_gathered["light-to-temperature map"] and not data_gathered["temperature-to-humidity map"]:
                    light_to_temperature.append((src, dest, range_length))

                # If the "temperature-to-humidity map" header has been passed, but not the "humidity-to-location map" header,
                # then the group of numbers belong to the "stemperature-to-humidity map".
                elif data_gathered["temperature-to-humidity map"] and not data_gathered["humidity-to-location map"]:
                    temperature_to_humidity.append((src, dest, range_length))

                # Else, append the group of numbers to the "humidity-to-location map".
                else:
                    humidity_to_location.append((src, dest, range_length))

        seed_location = [] # Store the location of all the seeds after the mappings

        # For seed in the list of seeds...
        for seed in seeds:
            # For each item in the mappings... (i.e., (source, destination, range))
            # If the first category's number is between the source and source + range numbers,
            # find the difference between the seed and source number and add the difference to the destination
            # to find the conversion between first and second categories' numbers.
            # Else, the the second category's number is equal to the first category's number.

            for i in range(len(seed_to_soil)):
                src, dest, range_length = label_mappings(seed_to_soil[i]) # Label the elements in the maps.

                if (seed >= src) and (seed < (src + range_length)):
                    soil = (seed - src) + dest
                    break
                else:
                    soil = seed
            
            for i in range(len(soil_to_fertililzer)):
                src, dest, range_length = label_mappings(seed_to_soil[i]) # Label the elements in the maps.
                if (soil >= src) and (soil < (src + range_length)):
                    fertilizer = (soil - src) + dest
                    break
                else:
                    fertilizer = soil

            for i in range(len(fertilizer_to_water)):
                src, dest, range_length = label_mappings(soil_to_fertililzer[i]) # Label the elements in the maps.
                if (fertilizer >= src) and (fertilizer < (src + range_length)):
                    water = (fertilizer - src) + dest
                    break
                else:
                    water = fertilizer

            for i in range(len(water_to_light)):
                src, dest, range_length = label_mappings(fertilizer_to_water[i]) # Label the elements in the maps.
                if (water >= src) and (water < (src + range_length)):
                    light = (water - src) + dest
                    break
                else:
                    light = water

            for i in range(len(light_to_temperature)):
                src, dest, range_length = label_mappings(water_to_light[i]) # Label the elements in the maps.
                if (light >= src) and (light < (src + range_length)):
                    temperature = (light - src) + dest
                    break
                else:
                    temperature = light

            for i in range(len(temperature_to_humidity)):
                src, dest, range_length = label_mappings(light_to_temperature[i]) # Label the elements in the maps.
                if (temperature >= src) and (temperature < (src + range_length)):
                    humidity = (temperature - src) + dest
                    break
                else:
                    humidity = temperature

            for i in range(len(humidity_to_location)):
                src, dest, range_length = label_mappings(temperature_to_humidity[i]) # Label the elements in the maps.
                if (humidity >= src) and (humidity < (src + range_length)):
                    location = (humidity - src) + dest
                    break
                else:
                    location = humidity

            # Add each location to the list of seed locations
            seed_location.append(location)

        # Print the lowest seed location
        print("Lowest Seed Location: ", min(seed_location))
        
if __name__ == "__main__":
    main()
```

```python
# Part 2
def getMapRange(input_range_start, input_range_end, map_conversions):
    new_ranges = [] # Store the new output ranges after the map conversions
    current_input_range_start = input_range_start # Save the original input range start
    
    # Find the first element where the input range start is less than or equal to the source range end 
    # to focus on splicing the input range there.
    # If it is no element is found to meet the criteria, return -1.
    element = next((r for r in map_conversions if input_range_start <= r[1]), -1)
    
    # If the element was found, find the index in the list of map conversions.
    # Otherwise, set the start index to -1.
    if element != -1:
        start_index = map_conversions.index(element)
    else:
        start_index = -1
    
    # If the start index is greater than or equal to zero, 
    # then the element was found in the list of map conversions.
    if start_index >= 0:
        # Iterate through the list of map conversions, starting from the start index.
        for i in range(start_index, len(map_conversions)):
            current_rule = map_conversions[i] # Look at the current source range

            # If the current input range start is less than the current source range start...
            if current_input_range_start < current_rule[0]:
                # If the input range end is less than the current range start as well,
                # the mapping of the input range is 1-to-1, so there is no change.
                if input_range_end < current_rule[0]:
                    new_ranges.append((current_input_range_start, input_range_end))
                    return new_ranges

                # Otherwise, split the input range to be from its original start to just before the source range start
                new_ranges.append((current_input_range_start, current_rule[0] - 1))

                # Update the current input range start to the current source range start
                current_input_range_start = current_rule[0]

            # If the input range end is less than or equal to the current source range end...
            if input_range_end <= current_rule[1]:
                # Convert the input range from the source start to the input range end to the new output range values by adding the offset
                # since the map conversion needs to be used.
                new_ranges.append((current_input_range_start + current_rule[2], input_range_end + current_rule[2]))
                return new_ranges
            
            # Otherwise, the entire input range is within the source mapping, so it needs to be converted to the destination
            # using the offset.
            new_ranges.append((current_input_range_start + current_rule[2], current_rule[1] + current_rule[2]))
            current_input_range_start = current_rule[1] + 1 # Update the current input range start to the source range end + 1 to exit the current rule
    
    # If the input range does not meet any of the above criteria, add the input range as is
    new_ranges.append((current_input_range_start, input_range_end))

    # Return the list of new ranges
    return new_ranges

def main():
    with open("input.txt", 'r') as input_file:
        # input_lines = [line.rstrip() for line in input_file.readlines() if line.rstrip() != ""] # Store string inputs into a list
        input_lines = [line.rstrip() for line in input_file.readlines()] # Store string inputs into a list
        
        # Get the seed range start and the length of the seed range
        seeds = [int(i) for i in input_lines[0].lstrip("seeds: ").split(" ")]
        seeds = [tuple(seeds[i:i+2]) for i in range(0, len(seeds), 2)]

        # Format the seed range to store the start and end, inclusively
        seed_ranges = []
        for seed in seeds:
            seed_ranges.append((seed[0], seed[0] + seed[1] - 1))
        seed_ranges.sort(key=lambda n: n[0])

        # Remove the first two items from the list of input lines since the seeds (and empty line) have been parsed
        input_lines = input_lines[2:]
        input_lines.append("")
        
        mappings = [] # Store the tuple of (source_start, source_end (inclusive), offset)
        # For each line in the list of input lines...
        for line_number in range(len(input_lines)):
            # If "map" is in the line, skip to the next line.
            if "map" in input_lines[line_number]:
                continue

            # If the line is empty, that means that the end of the mapping category has been reached.
            # Convert the current range to the next range.
            if input_lines[line_number] == "":
                mappings.sort(key=lambda n: n[0]) # Sort the mappings to make it easier to reformat the input ranges

                output_ranges = [] # Clear and store the output ranges after the map conversions
                # For each seed (or input range) in the list of seeds...
                for seed in seed_ranges:
                    range_start = seed[0]
                    range_end = seed[1]

                    # After reformatting the new ranges (or keeping them the same if they were 1-to-1 mappings), 
                    # add the new ranges to the list of output ranges.
                    output_ranges.extend(getMapRange(range_start, range_end, mappings))
                
                seed_ranges = output_ranges
                mappings.clear()

            # If the line begins with a digit, then store the (destination, source, range) into a mapping list.
            if any(char.isdigit() for char in input_lines[line_number]):
                dest, src, range_length = [int(i) for i in input_lines[line_number].split(" ")] # Get the destination, source, and range from the line

                # Since the destination range can be calculated from the offset between the source and destination start differences,
                # reformat the tuple as (source_start, source_end (inclusive), offset).
                offset = dest - src
                mappings.append((src, src + range_length - 1, offset))

        seed_ranges.sort(key=lambda n: n[0])

        # Print the lowest location number.
        print("Lowest Location Number: ", seed_ranges[0][0])
        
if __name__ == "__main__":
    main()
```
